```python
import numpy as np
import pandas as pd

#import datatable
from sklearn.preprocessing import LabelEncoder
import re
import seaborn as sns
import matplotlib.pyplot as plt
from fbprophet import Prophet
```

```python
train_df=pd.read_csv('train_2.csv')
```

```python
def features(page, n):
    a = re.search('(.*)\_(.*).wikipedia.org\_(.*)\_(.*)', page)
    if a is not None:
        if (n==1):
            return a.group(1)
        elif (n==2):
            return a.group(2)
        elif (n==3):
            return a.group(3)
        elif (n==4):
            return a.group(4)
    else:
        return ""   
    
train_df['page_name'] = train_df['Page'].apply(lambda x: features(x, 1))
train_df['lang'] = train_df['Page'].apply(lambda x: features(x, 2))
train_df['version'] = train_df['Page'].apply(lambda x: features(x, 3))
train_df['spider'] = train_df['Page'].apply(lambda x: features(x, 4))

train_df['sum'] = train_df.sum(axis=1)
```
